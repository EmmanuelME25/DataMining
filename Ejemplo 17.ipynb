### Sigmoid

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
x= np.linspace(-10,10,100)
y = sigmoid(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.title("Sigmoide")
plt.xlabel("X")
plt.ylabel("sigmoid(x)")
plt.grid(True)
plt.show()
```

### Tangente hiperb贸lica

```python
def tanh(x):
    return (2/(1+np.exp(-2*x)))-1
```

```python
y1 = tanh(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y1)
plt.title("Tanh")
plt.xlabel("X")
plt.ylabel("Tanh(x)")
plt.grid(True)
plt.show()
```

### Rectified Linear Unit (ReLu)

```python
def relu(x):
    return np.maximum(0,x)
```

```python
y3 = relu(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y3)
plt.title("Relu")
plt.xlabel("X")
plt.ylabel("Relu(x)")
plt.grid(True)
plt.show()
```

### Leaky ReLU

```python
def lerelu(x):
    return np.maximum(0.25*x,x)
```

```python
y4 = lerelu(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y4)
plt.title("Leaky Relu")
plt.xlabel("X")
plt.ylabel("Leaky ReLU(x)")
plt.grid(True)
plt.show()
```

### Exponential Linear Unit (ELU)

```python
def elu(x):
    return np.where(x > 0,x, 0.5*(np.exp(x)-1))
```

```python
y5 = elu(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y5)
plt.title("ELU")
plt.xlabel("X")
plt.ylabel("ELU(x)")
plt.grid(True)
plt.show()
```

### Softmax

```python
def sof(x):
    exp_x = np.exp(x - np.max(x))
    return exp_x/exp_x.sum(axis=0, keepdims=True)
```

```python
y6 =sof(x)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y6)
plt.title("Softmax")
plt.xlabel("X")
plt.ylabel("Softmax(x)")
plt.grid(True)
plt.show()
```
