## Ejemplo 26

### Librerias

```python
from sklearn.datasets import load_iris
from sklearn import tree
```

### Cargar datos

```python
iris = load_iris()
```

### Dividir datos objetivo

```python
X, y = iris.data, iris.target
```

### Clasificador

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X, y)
```

```python
tree.plot_tree(clf)
```
